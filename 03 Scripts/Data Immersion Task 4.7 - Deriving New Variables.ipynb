#### 2. Creating price label definitions

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

#### 3. Applying price label to new data column

```python
df_subset['price_range'] = df_subset.apply(price_label, axis=1)
```

```python
df_subset.drop(columns=['_merge', 'price range'], inplace=True)
```

#### 4. Applying label for order frequency on day of week "busiest day"

```python
df['orders_day_of_week'].value_counts(dropna = False)
```

```python
result = []

for value in df["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
df['busiest day'] = result
```

```python
df['busiest day'].value_counts(dropna = False)
```

#### 1. Making a "busiest days" column

```python
result = []

for value in df["orders_day_of_week"]:
  if (value == 0) or (value == 1):
    result.append("Busiest days")
  elif (value == 4) or (value == 3):
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
df['busiest days'] = result
```

#### 2. Find the times (hours) that are most busy, average, and least busy and create a new column

```python
df['order_hour_of_day'].value_counts(dropna = False)
```

```python
result = []

for value in df["order_hour_of_day"]:
  if (value >= 10) and (value <= 15):
    result.append("Most orders")
  elif (value >= 0) and (value <= 6):
    result.append("Fewest orders")
  else:
    result.append("Average orders")
```

```python
result
```

```python
df['busiest_period_of_day'] = result
```
