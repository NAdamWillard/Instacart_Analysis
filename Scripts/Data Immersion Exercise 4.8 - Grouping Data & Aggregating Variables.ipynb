#### 2. Grouping by various means

```python
# Grouping by product name

df_subset.groupby('product_name')
```

```python
# Determining mean of order numbers by department

df_subset.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# an alternative method not using agg function

df_subset.groupby('department_id')['order_number'].mean()
```

```python
# more agg statistics (at the same time)

df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

#### 3. Determining customer loyalty based on max order numbers

```python
df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df.head(15)
```

```python
pd.options.display.max_rows = None
```

```python
# Assigning loyalty flag in new column

df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df.loc[(df['max_order'] <= 40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'Newer customer'
```
